### Exercise 1: Read Total profit of all months and show it using a line plot.

```python
# total profit of all months and show line plot.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.figure(figsize=(10, 6))
plt.plot(months, data['total_profit'], marker='*', color='red', linewidth=1, label='Profit of last year')
plt.title('Company profit per month')
plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.grid(True)
plt.legend()
plt.show()
```

### Exercise 2: Read data for Bathing soap and facewash of all months and display it using the Subplot.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

#Create a figure and subplots of 1 rows,  2 column
fig, axs = plt.subplots(1, 2, figsize=(20, 6))

#Ploting data for bathingsoap in the first subplot
axs[0].plot(months, data['bathingsoap'], marker='*', color='blue', label='Bathing Soap')
axs[0].set_title('Monthly Sales of Bathing Soap')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Sales')
axs[0].grid(True)
axs[0].legend()

#Plot data for facewash in the second subplot
axs[1].plot(months, data['facewash'], marker='*', color='green', label='Facewash')
axs[1].set_title('Monthly Sales of Facewash')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Sales')
axs[1].grid(True)
axs[1].legend()

#Show the plot
plt.show()
```
